```python
# 4.3.7 –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö

# C—Ç–æ–ª–±—Ü—ã —Å –¥–∞—Ç–æ–π/–≤—Ä–µ–º–µ–Ω–µ–º –≤ datetime
orders_df['created_dt_msk'] = pd.to_datetime(orders_df['created_dt_msk'])
orders_df['created_ts_msk'] = pd.to_datetime(orders_df['created_ts_msk'])

# –ü–æ–Ω–∏–∂–∞–µ–º —Ä–∞–∑—Ä—è–¥–Ω–æ—Å—Ç—å –¥–ª—è revenue –∏ total
float_cols = ['revenue', 'total']
for col in float_cols:
    orders_df[col] = pd.to_numeric(orders_df[col], downcast='float')
```

```python
# 4.5.2 –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–∞–∫–∞–∑–æ–≤ –±–∏–ª–µ—Ç–æ–≤ –¥–ª—è –ª–µ—Ç–Ω–µ–≥–æ –∏ –æ—Å–µ–Ω–Ω–µ–≥–æ –ø–µ—Ä–∏–æ–¥–æ–≤

# –û–±—ä–µ–¥–∏–Ω—è–µ–º –¥–∞—Ç–∞—Å–µ—Ç—ã orders_df –∏ events_df –ø–æ event_id
df = orders_df.merge(
    events_df[['event_id', 'event_type_main']],
    on='event_id',
    how='left'
)

# –î–æ–±–∞–≤–ª—è–µ–º month –∏ season (—Ç–æ–ª—å–∫–æ –ª–µ—Ç–æ –∏ –æ—Å–µ–Ω—å)
df['month'] = df['created_dt_msk'].dt.month
df['season'] = df['month'].map(lambda m: '–ª–µ—Ç–æ' if m in (6,7,8)
                               else '–æ—Å–µ–Ω—å' if m in (9,10,11)
                               else None)
df = df[df['season'].isin(['–ª–µ—Ç–æ','–æ—Å–µ–Ω—å'])]

# –§—É–Ω–∫—Ü–∏—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –¥–æ–ª–µ–π
def plot_prop(col, title):
    cnt = df.groupby(['season', col]).size().unstack(fill_value=0)
    prop = cnt.div(cnt.sum(axis=1), axis=0)
    prop.T.plot(kind='bar', figsize=(10,5))
    plt.title(title)
    plt.ylabel('–î–æ–ª—è –∑–∞–∫–∞–∑–æ–≤')
    plt.xlabel(col)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='–°–µ–∑–æ–Ω')
    plt.tight_layout()
    plt.show()

# 5. –ì—Ä–∞—Ñ–∏–∫–∏
plot_prop('event_type_main',       '–î–æ–ª–∏ –∑–∞–∫–∞–∑–æ–≤ –ø–æ —Ç–∏–ø—É –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è')
plot_prop('device_type_canonical', '–î–æ–ª–∏ –∑–∞–∫–∞–∑–æ–≤ –ø–æ —Ç–∏–ø—É —É—Å—Ç—Ä–æ–π—Å—Ç–≤–∞')
plot_prop('age_limit',             '–î–æ–ª–∏ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–º—É —Ä–µ–π—Ç–∏–Ω–≥—É')
```

```python
# –û–±—ä–µ–¥–∏–Ω—è–µ–º –¥–∞—Ç–∞—Å–µ—Ç—ã orders_df –∏ events_df –ø–æ event_id
orders_full = orders_df.merge(
    events_df[['event_id', 'event_type_main']],
    on='event_id',
    how='left'
)

# –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –Ω–∞ –ª–µ—Ç–æ –∏ –æ—Å–µ–Ω—å
df_seasons = orders_full[orders_full['season'].isin(['–ª–µ—Ç–æ', '–æ—Å–µ–Ω—å'])]

# 3. –°—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω—é—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –±–∏–ª–µ—Ç–∞ –ø–æ —Ç–∏–ø—É –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è –∏ —Å–µ–∑–æ–Ω—É
mean_rev = (
    df_seasons
    .groupby(['event_type_main', 'season'])['one_ticket_revenue_rub']
    .mean()
    .unstack()
    .rename(columns={'–ª–µ—Ç–æ': 'summer_avg', '–æ—Å–µ–Ω—å': 'autumn_avg'})
)

# –í—ã—á–∏—Å–ª—è–µ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –æ—Å–µ–Ω—å—é
mean_rev['pct_change'] = (mean_rev['autumn_avg'] / mean_rev['summer_avg'] - 1) * 100

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.figure(figsize=(10, 5))
mean_rev['pct_change'].sort_values().plot(kind='bar')
plt.axhline(0, color='gray', linewidth=0.8)
plt.title('–ò–∑–º–µ–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –±–∏–ª–µ—Ç–∞ (–æ—Å–µ–Ω—å vs –ª–µ—Ç–æ), %')
plt.ylabel('–û—Ç–Ω–æ—à–µ–Ω–∏–µ –æ—Å–µ–Ω–Ω–µ–≥–æ –ø–µ—Ä–∏–æ–¥–∞ –∫ –ª–µ—Ç–Ω–µ–º—É')
plt.xlabel('–¢–∏–ø –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# –ò—Ç–æ–≥–æ–≤–∞—è —Ç–∞–±–ª–∏—Ü–∞
print(mean_rev[['summer_avg', 'autumn_avg', 'pct_change']])
```

- –í–æ –≤—Å–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏—è—Ö (–¥—Ä—É–≥–æ–µ, –∫–æ–Ω—Ü–µ—Ä—Ç—ã, —Å–ø–æ—Ä—Ç, —Ç–µ–∞—Ç—Ä, —ë–ª–∫–∏) —Å—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ —É–ø–∞–ª–∞, –ø—Ä–∏—á—ë–º —Å–∞–º—ã–π —Å–∏–ª—å–Ω—ã–π —Å–ø–∞–¥ –≤ —Ç–µ–∞—Ç—Ä–µ (-17,8%) –∏ –Ω–∞ "—ë–ª–∫–∞—Ö" (-15,2%).

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –ù–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω—ã —Ä–∞—Å—á–µ—Ç—ã –¥–ª—è `device_type_canonical` –∏ `age_limit`</div>

```python
# –î–æ–ª–∏ –ø–æ device_type_canonical –≤ —Å–µ–∑–æ–Ω–Ω—ã—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –∏ –∏—Ö –∏–∑–º–µ–Ω–µ–Ω–∏–µ (–æ—Å–µ–Ω—å vs –ª–µ—Ç–æ)

dev_counts = (
    df_seasons
    .groupby(['device_type_canonical', 'season'])
    .size()
    .unstack(fill_value=0)
)

# –ü–µ—Ä–µ–≤–æ–¥–∏–º –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω—ã–µ –¥–æ–ª–∏ –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ —Å–µ–∑–æ–Ω–∞
dev_shares = dev_counts.div(dev_counts.sum(axis=0), axis=1) * 100

# –°—á–∏—Ç–∞–µ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –¥–æ–ª–∏ –æ—Å–µ–Ω—å—é –∫ –ª–µ—Ç—É
dev_shares['diff_pp'] = dev_shares['–æ—Å–µ–Ω—å'] - dev_shares['–ª–µ—Ç–æ']

print(dev_shares)
```

–ö–û–ú–ú–ï–ù–¢–ê–†–ò–ô: –î–æ–ª—è –¥–µ—Å–∫—Ç–æ–ø-–∑–∞–∫–∞–∑–æ–≤ –≤—ã—Ä–æ—Å–ª–∞ —Å 19,5% –ª–µ—Ç–æ–º –¥–æ 20,4% –æ—Å–µ–Ω—å—é (+0,9 –ø.–ø.), —Ç–æ–≥–¥–∞ –∫–∞–∫ –º–æ–±–∏–ª—å–Ω—ã–µ –∑–∞–∫–∞–∑—ã —Å–æ–∫—Ä–∞—Ç–∏–ª–∏—Å—å —Å 80,5% –¥–æ 79,6% (-0.9 –ø.–ø.).

```python
# –î–æ–ª–∏ –ø–æ age_limit –≤ —Å–µ–∑–æ–Ω–Ω—ã—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –∏ –∏—Ö –∏–∑–º–µ–Ω–µ–Ω–∏–µ (–æ—Å–µ–Ω—å vs –ª–µ—Ç–æ)


age_counts = (
    df_seasons
    .groupby(['age_limit', 'season'])
    .size()
    .unstack(fill_value=0)
)

# –ü–µ—Ä–µ–≤–æ–¥–∏–º –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω—ã–µ –¥–æ–ª–∏ –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ —Å–µ–∑–æ–Ω–∞
age_shares = age_counts.div(age_counts.sum(axis=0), axis=1) * 100

# –°—á–∏—Ç–∞–µ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –¥–æ–ª–∏ –æ—Å–µ–Ω—å—é –∫ –ª–µ—Ç—É
age_shares['diff_pp'] = age_shares['–æ—Å–µ–Ω—å'] - age_shares['–ª–µ—Ç–æ']

print(age_shares)
```

```python
# 4.6.1 –î–∏–Ω–∞–º–∏–∫–∞ –∏–∑–º–µ–Ω–µ–Ω–∏–π –ø–æ –¥–Ω—è–º —Å –ª–∏–Ω–∏—è–º–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –∏ –º–µ–¥–∏–∞–Ω—ã

autumn_df = orders_df.loc[orders_df['created_dt_msk'].dt.month.isin([9, 10])].copy()
autumn_df['order_date'] = autumn_df['created_dt_msk'].dt.date

# —Å–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞
pivot = autumn_df.groupby('order_date').agg(
    orders_count     = ('order_id',               'count'),
    dau              = ('user_id',                'nunique'),
    avg_ticket_price = ('one_ticket_revenue_rub', 'mean')
).reset_index()

# —Å—Ä–µ–¥–Ω–∏–µ –∑–∞–∫–∞–∑—ã –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
pivot['avg_orders_per_user'] = pivot['orders_count'] / pivot['dau']

# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
fig, axes = plt.subplots(4, 1, figsize=(12, 16), sharex=False)

metrics = [
    ('orders_count', '–û–±—â–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ (–µ–∂–µ–¥–Ω–µ–≤–Ω–æ)'),
    ('dau', 'DAU: –∞–∫—Ç–∏–≤–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏'),
    ('avg_orders_per_user', '–°—Ä–µ–¥–Ω–∏–µ –∑–∞–∫–∞–∑—ã –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è'),
    ('avg_ticket_price','–°—Ä–µ–¥–Ω—è—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –±–∏–ª–µ—Ç–∞ (—Ä—É–±.)')
]

for ax, (col, title) in zip(axes, metrics):
    # –ª–∏–Ω–∏—è –º–µ—Ç—Ä–∏–∫
    ax.plot(pivot['order_date'], pivot[col], marker='o')
    ax.set_title(title)
    ax.tick_params(axis='x', rotation=45, labelbottom=True)
    ax.set_xlabel('–î–∞—Ç–∞')

    # —Å—Ä–µ–¥–Ω–µ–µ
    mean_val = pivot[col].mean()
    ax.axhline(mean_val, linestyle='--', linewidth=1, label=f'–°—Ä–µ–¥–Ω–µ–µ ({mean_val:.2f})')

    # –º–µ–¥–∏–∞–Ω–∞
    med_val = pivot[col].median()
    ax.axhline(med_val, color='gray', linestyle=':', linewidth=1, label=f'–ú–µ–¥–∏–∞–Ω–∞ ({med_val:.2f})')

    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# 4.6.2 –ù–µ–¥–µ–ª—å–Ω–∞—è —Ü–∏–∫–ª–∏—á–Ω–æ—Å—Ç—å 

# –î–æ–±–∞–≤–ª—è–µ–º –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –∏ —Ñ–ª–∞–≥ –≤—ã—Ö–æ–¥–Ω–æ–≥–æ
orders_df['order_date']  = orders_df['created_dt_msk'].dt.date
orders_df['weekday_num'] = orders_df['created_dt_msk'].dt.weekday 
orders_df['is_weekend']  = orders_df['weekday_num'] >= 5

# –î–æ–±–∞–≤–ª—è–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –¥–Ω–µ–π
weekday_map = {
    0: '–ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫', 1: '–≤—Ç–æ—Ä–Ω–∏–∫', 2: '—Å—Ä–µ–¥–∞',
    3: '—á–µ—Ç–≤–µ—Ä–≥', 4: '–ø—è—Ç–Ω–∏—Ü–∞', 5: '—Å—É–±–±–æ—Ç–∞', 6: '–≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ'
}
orders_df['weekday'] = orders_df['weekday_num'].map(weekday_map)

# –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ –∫–∞–∂–¥–æ–º—É –¥–Ω—é –Ω–µ–¥–µ–ª–∏
weekly = orders_df.groupby('weekday').agg(
    orders_count     = ('order_id', 'count'),
    dau              = ('user_id', 'nunique'),
    avg_ticket_price = ('one_ticket_revenue_rub', 'mean')
).reindex([
    '–ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫','–≤—Ç–æ—Ä–Ω–∏–∫','—Å—Ä–µ–¥–∞','—á–µ—Ç–≤–µ—Ä–≥','–ø—è—Ç–Ω–∏—Ü–∞','—Å—É–±–±–æ—Ç–∞','–≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ'
])

# –î–æ–±–∞–≤–ª—è–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–∞–∫–∞–∑—ã –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
weekly['avg_orders_user'] = weekly['orders_count'] / weekly['dau']

print("–ú–µ—Ç—Ä–∏–∫–∏ –ø–æ –∫–∞–∂–¥–æ–º—É –¥–Ω—é –Ω–µ–¥–µ–ª–∏:")
print(weekly)

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –±—É–¥–Ω–∏—Ö –∏ –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–Ω–µ–π
daily = orders_df.groupby('is_weekend').agg(
    orders_count     = ('order_id',               'count'),
    dau              = ('user_id',                'nunique'),
    avg_ticket_price = ('one_ticket_revenue_rub', 'mean')
).rename(index={False:'–ë—É–¥–Ω–∏', True:'–í—ã—Ö–æ–¥–Ω—ã–µ'})

# –î–æ–±–∞–≤–ª—è–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–∞–∫–∞–∑—ã –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
daily['avg_orders_user'] = daily['orders_count'] / daily['dau']

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –±—É–¥–Ω–∏ vs –≤—ã—Ö–æ–¥–Ω—ã–µ:")
print(daily)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –±—É–¥–Ω–∏/–≤—ã—Ö–æ–¥–Ω—ã–µ
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
metrics = ['orders_count', 'dau', 'avg_orders_user', 'avg_ticket_price']
titles  = ['–ß–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤','DAU','–°—Ä–µ–¥–Ω–∏–µ –∑–∞–∫–∞–∑—ã –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è','–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –±–∏–ª–µ—Ç–∞']

for ax, m, t in zip(axes.flatten(), metrics, titles):
    daily[m].plot(kind='bar', ax=ax, legend=False)
    ax.set_title(t)
    ax.set_xticklabels(daily.index, rotation=0)
    ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# 4.7.1 –ú–µ—Ç—Ä–∏–∫–∏ –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º
# –û–±—ä–µ–¥–∏–Ω—è–µ–º –∑–∞–∫–∞–∑—ã —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ —Ä–µ–≥–∏–æ–Ω–µ –∏–∑ events_df
orders_events = orders_df.merge(
    events_df[['event_id', 'region_name']],
    on='event_id',
    how='left'
)

# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ —Ä–µ–≥–∏–æ–Ω—É
region_stats = orders_events.groupby('region_name').agg(
    unique_events=('event_id', 'nunique'),
    orders_count  =('order_id', 'count')
).reset_index()

# –î–æ–ª–∏ 
total_events = region_stats['unique_events'].sum()
total_orders = region_stats['orders_count'].sum()
region_stats['events_share'] = region_stats['unique_events'] / total_events
region_stats['orders_share'] = region_stats['orders_count']    / total_orders

# –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —á–∏—Å–ª—É —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π
region_stats = region_stats.sort_values('unique_events', ascending=False)

display(region_stats)
```

```python
user_orders = (
    orders_df
    .groupby(['user_id', 'device_type_canonical'])
    .agg(orders_per_user=('order_id', 'count'))
    .reset_index()
)

# —Å–æ–∑–¥–∞—ë–º —Å–µ—Ä–∏–∏ mobile –∏ desktop
mobile = user_orders.loc[user_orders['device_type_canonical']=='mobile', 'orders_per_user']
desktop = user_orders.loc[user_orders['device_type_canonical']=='desktop', 'orders_per_user']

# –≤—ã—á–∏—Å–ª—è–µ–º —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è Shapiro
n_mob = min(len(mobile), 5000)
n_des = min(len(desktop), 5000)

# —Ç–µ—Å—Ç –®–∞–ø–∏—Ä–æ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å
shapiro_mobile  = stats.shapiro(mobile.sample(n_mob, random_state=1))
shapiro_desktop = stats.shapiro(desktop.sample(n_des, random_state=1))

print("Shapiro mobile:",  shapiro_mobile)
print("Shapiro desktop:", shapiro_desktop)

# –õ–µ–≤–µ–Ω–µ —É—Å—Ç–æ–π—á–∏–≤ –∫ –Ω–µ–Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏, –ø–æ—ç—Ç–æ–º—É –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤—Å–µ –¥–∞–Ω–Ω—ã–µ
levene_stat, levene_p = stats.levene(
    mobile,
    desktop,
    center='median'
)

print(f"Levene statistic: {levene_stat:.3f}, p-value: {levene_p:.3f}")
```

- –¢–µ—Å—Ç –õ–µ–≤–µ–Ω–µ
—Ä-value = 0,004 < 0,05, —á—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –º—ã –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –¥–∏—Å–ø–µ—Ä—Å–∏–π –≤ –≥—Ä—É–ø–ø–∞—Ö. –î–∏—Å–ø–µ—Ä—Å–∏–∏ —á–∏—Å–ª–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –º–æ–±–∏–ª—å–Ω–æ–π –∏ –¥–µ—Å–∫—Ç–æ–ø–Ω–æ–π –≥—Ä—É–ø–ø–∞—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è.

–ü–æ—Å–∫–æ–ª—å–∫—É –Ω–∞—Ä—É—à–µ–Ω—ã –æ–±–∞ –∫–ª—é—á–µ–≤—ã—Ö —É—Å–ª–æ–≤–∏—è –¥–ª—è –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–≥–æ t-—Ç–µ—Å—Ç–∞ (–Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å –∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –¥–∏—Å–ø–µ—Ä—Å–∏–π), –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–∏—Ö –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å **–Ω–µ–ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π U-—Ç–µ—Å—Ç –ú–∞–Ω–Ω–∞‚Äì–£–∏—Ç–Ω–∏.**

```python
# –û–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π U-—Ç–µ—Å—Ç (mobile > desktop)
u_stat, p_value = stats.mannwhitneyu(mobile, desktop, alternative='greater')

#  Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —ç—Ñ—Ñ–µ–∫—Ç–∞
n1, n2 = len(mobile), len(desktop)
mean_U = n1 * n2 / 2
std_U = np.sqrt(n1 * n2 * (n1 + n2 + 1) / 12)
z = (u_stat - mean_U) / std_U

print(f"U-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {u_stat:.1f}")
print(f"p-value: {p_value:.4f}")
print(f"Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {z:.3f}")
```

**–í–´–í–û–î: –ú—ã –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –∏ –ø—Ä–∏–Ω–∏–º–∞–µ–º, —á—Ç–æ –º–æ–±–∏–ª—å–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –¥–µ–ª–∞—é—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º –±–æ–ª—å—à–µ –∑–∞–∫–∞–∑–æ–≤, —á–µ–º –¥–µ—Å–∫—Ç–æ–ø–Ω—ã–µ.**

4.8.2 –ì–∏–ø–æ—Ç–µ–∑–∞ –æ —Å—Ä–µ–¥–Ω–µ–º –≤—Ä–µ–º–µ–Ω–∏ –º–µ–∂–¥—É –∑–∞–∫–∞–∑–∞–º–∏ 

> –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã **"–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –º–µ–∂–¥—É –∑–∞–∫–∞–∑–∞–º–∏ —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –º–æ–±–∏–ª—å–Ω—ã—Ö —É—Å—Ç—Ä–æ–π—Å—Ç–≤ –≤—ã—à–µ, —á–µ–º —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–µ—Å–∫—Ç–æ–ø–∞"**

>> –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0): —Å—Ä–µ–¥–Ω–∏–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –º–µ–∂–¥—É –ø–æ–∫—É–ø–∫–∞–º–∏ (days_since_prev) –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π mobile —Ä–∞–≤–µ–Ω —Å—Ä–µ–¥–Ω–µ–º—É –∏–Ω—Ç–µ—Ä–≤–∞–ª—É –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π desktop.

>> –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (**H1**): —Å—Ä–µ–¥–Ω–∏–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –º–µ–∂–¥—É –ø–æ–∫—É–ø–∫–∞–º–∏ —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π mobile –±–æ–ª—å—à–µ, —á–µ–º —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π desktop.

> –¢–∏–ø —Ç–µ—Å—Ç–∞

>> –ü–æ—Å–∫–æ–ª—å–∫—É –º—ã —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –¥–≤–µ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –≤—ã–±–æ—Ä–∫–∏ –∏ –∏–Ω—Ç–µ—Ä–µ—Å—É–µ–º—Å—è —Ç–æ–ª—å–∫–æ —Ç–µ–º, –ø—Ä–µ–≤—ã—à–∞–µ—Ç –ª–∏ —Å—Ä–µ–¥–Ω–µ–µ –≤ –æ–¥–Ω–æ–π –≥—Ä—É–ø–ø–µ —Å—Ä–µ–¥–Ω–µ–µ –≤ –¥—Ä—É–≥–æ–π, –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ç–µ—Å—Ç.
>> –ü–µ—Ä–µ–¥ —ç—Ç–∏–º –ø—Ä–æ–≤–µ—Ä–∏–º, –ø–æ–¥—Ö–æ–¥—è—Ç –ª–∏ –¥–∞–Ω–Ω—ã–µ –¥–ª—è t-—Ç–µ—Å—Ç–∞ –°—Ç—å—é–¥–µ–Ω—Ç–∞ (—Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ —Å—Ö–æ–∂–µ—Å—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏–π).
>> –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω—ã –∏–ª–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è, –ø–µ—Ä–µ–∫–ª—é—á–∞–µ–º—Å—è –Ω–∞ –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –Ω–µ–ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π U-—Ç–µ—Å—Ç –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏.


```python
autumn = orders_df.loc[
    (orders_df['season'] == '–æ—Å–µ–Ω—å') &
    (orders_df['days_since_prev'].notna())
].copy()

# 2. –°–µ—Ä–∏–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –ø–æ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–∞–º
mobile_days  = autumn.loc[autumn['device_type_canonical']=='mobile',  'days_since_prev']
desktop_days = autumn.loc[autumn['device_type_canonical']=='desktop', 'days_since_prev']

# 3. –¢–µ—Å—Ç –®–∞–ø–∏—Ä–æ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å (–¥–æ 5000 —Å–ª—É—á–∞–π–Ω—ã—Ö –Ω–∞–±–ª—é–¥–µ–Ω–∏–π)
n_mob = min(len(mobile_days), 5000)
n_des = min(len(desktop_days), 5000)
print("Shapiro mobile:",  stats.shapiro(mobile_days.sample(n_mob, random_state=1)))
print("Shapiro desktop:", stats.shapiro(desktop_days.sample(n_des, random_state=1)))

# 4. –¢–µ—Å—Ç –õ–µ–≤–µ–Ω–µ –Ω–∞ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –¥–∏—Å–ø–µ—Ä—Å–∏–π
lev_stat, lev_p = stats.levene(mobile_days, desktop_days, center='median')
print(f"Levene p-value: {lev_p:.3f}")
```

- –¢–µ—Å—Ç –õ–µ–≤–µ–Ω–µ
—Ä-value = 0,000 < 0,05, —á—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –º—ã –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –¥–∏—Å–ø–µ—Ä—Å–∏–π –≤ –≥—Ä—É–ø–ø–∞—Ö. –î–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—Ä–µ–º–µ–Ω–∏ –∑–∞–∫–∞–∑–æ–≤ –≤ –º–æ–±–∏–ª—å–Ω–æ–π –∏ –¥–µ—Å–∫—Ç–æ–ø–Ω–æ–π –≥—Ä—É–ø–ø–∞—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è.

–ü–æ—Å–∫–æ–ª—å–∫—É –Ω–∞—Ä—É—à–µ–Ω—ã –æ–±–∞ –∫–ª—é—á–µ–≤—ã—Ö —É—Å–ª–æ–≤–∏—è –¥–ª—è –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–≥–æ t-—Ç–µ—Å—Ç–∞ (–Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å –∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –¥–∏—Å–ø–µ—Ä—Å–∏–π), –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–∏—Ö –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å **–Ω–µ–ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π U-—Ç–µ—Å—Ç –ú–∞–Ω–Ω–∞‚Äì–£–∏—Ç–Ω–∏.**

```python
u_stat, p_value = stats.mannwhitneyu(
    mobile_days,
    desktop_days,
    alternative='greater'
)

#  Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —ç—Ñ—Ñ–µ–∫—Ç–∞
n1, n2 = len(mobile), len(desktop)
mean_U = n1 * n2 / 2
std_U = np.sqrt(n1 * n2 * (n1 + n2 + 1) / 12)
z = (u_stat - mean_U) / std_U

print(f"U-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {u_stat:.1f}")
print(f"p-value: {p_value:.4f}")
print(f"Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {z:.3f}")
```

- –ò–∑—É—á–∏—Ç–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –¥–∞–Ω–Ω—ã—Ö: –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ, –∏–∑—É—á–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç–µ –≥–∏–ø–æ—Ç–µ–∑—ã –∏—Ö –ø–æ—è–≤–ª–µ–Ω–∏—è. –ü—Ä–æ–≤–µ–¥–∏—Ç–µ –æ–±—Ä–∞–±–æ—Ç–∫—É –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: –≤—ã –º–æ–∂–µ—Ç–µ –∑–∞–º–µ–Ω–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, —É–¥–∞–ª–∏—Ç—å —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –∏–ª–∏ –æ—Å—Ç–∞–≤–∏—Ç—å –∏—Ö –∫–∞–∫ –µ—Å—Ç—å.

```python
df['hours'] = df['hours'].fillna("Unknown")

#—Ç–∞–∫ –∫–∞–∫ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö —É–∂–µ –µ—Å—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏—è "unknown" —Ç–æ –¥–æ–±–∞–≤–ª—è–µ–º —ç—Ç—É –∫–∞—Ç–µ–≥–æ—Ä–∏—é –≤ —Ç–µ —è—á–µ–π–∫–∏ –≤ –∫–æ—Ç–æ—Ä—ã—Ö –µ–µ –Ω–µ—Ç
for col in ['price', 'avg_bill']:
    if col in df.columns and df[col].dtype.name == 'category':
        if "Unknown" not in df[col].cat.categories:
            df[col] = df[col].cat.add_categories("Unknown")
        df[col] = df[col].fillna("Unknown")
```

- –°—Ç–æ–ª–±—Ü—ã 'hours', 'price' –∏ 'avg_bill' —Å–æ–¥–µ—Ä–∂–∞—Ç —Ç–µ–∫—Å—Ç–æ–≤—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é, –∫–æ—Ç–æ—Ä—É—é –Ω–µ –≤—Å–µ–≥–¥–∞ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è—é—Ç, –ø–æ—ç—Ç–æ–º—É –ø—Ä–æ–ø—É—Å–∫–∏ –º–æ–≥—É—Ç –ø–æ—è–≤–ª—è—Ç—å—Å—è –ø—Ä–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–∏ —ç—Ç–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏. –ó–∞–ø–æ–ª–Ω–∏–º –∏—Ö –∑–Ω–∞—á–µ–Ω–∏–µ–º "Unknown".

```python
seats_median = df['seats'].median()
df['seats'] = df['seats'].fillna(seats_median)
```
